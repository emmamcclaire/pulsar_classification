```python
avgs = '''
SELECT AVG(itg_mean) AS avg_itg_mean,
    AVG(itg_stdev) AS avg_itg_stdev,
    AVG(itg_kurtosis) AS avg_itg_kurtosis,
    AVG(itg_skewness) AS avg_itg_skewness,
    AVG(dm_snr_mean) AS avg_dm_snr_mean,
    AVG(dm_snr_stdev) AS avg_dm_snr_stdev,
    AVG(dm_snr_kurtosis) AS avg_dm_snr_kurtosis,
    AVG(dm_snr_skew) AS avg_dm_snr_skewness
FROM htru2
GROUP BY candidate_pulsar
HAVING candidate_pulsar = 1
'''

avgs_df = pd.read_sql(avgs, engine)

avgs_df.head()
```

```python
avgs = '''
WITH pulsar_avgs AS (
SELECT 
    AVG(itg_mean) AS avg_itg_mean,
    AVG(itg_stdev) AS avg_itg_stdev,
    AVG(itg_kurtosis) AS avg_itg_kurtosis,
    AVG(itg_skewness) AS avg_itg_skewness,
    AVG(dm_snr_mean) AS avg_dm_snr_mean,
    AVG(dm_snr_stdev) AS avg_dm_snr_stdev,
    AVG(dm_snr_kurtosis) AS avg_dm_snr_kurtosis,
    AVG(dm_snr_skew) AS avg_dm_snr_skewness
FROM htru2
GROUP BY candidate_pulsar
HAVING candidate_pulsar = 1
)
SELECT htru2.itg_mean - pulsar_avgs.avg_itg_mean AS "diff_from_mean"
FROM htru2, pulsar_avgs
'''

avgs_df = pd.read_sql(avgs, engine)

avgs_df.head()
```

```python
avgs = '''
WITH pulsar_avgs AS (
SELECT 
    AVG(itg_mean) AS avg_itg_mean,
    AVG(itg_stdev) AS avg_itg_stdev,
    AVG(itg_kurtosis) AS avg_itg_kurtosis,
    AVG(itg_skewness) AS avg_itg_skewness,
    AVG(dm_snr_mean) AS avg_dm_snr_mean,
    AVG(dm_snr_stdev) AS avg_dm_snr_stdev,
    AVG(dm_snr_kurtosis) AS avg_dm_snr_kurtosis,
    AVG(dm_snr_skew) AS avg_dm_snr_skewness
FROM htru2
GROUP BY candidate_pulsar
HAVING candidate_pulsar = 1
), squared_diff AS (
SELECT htru2.itg_mean, POWER(htru2.itg_mean - pulsar_avgs.avg_itg_mean, 2) AS "squared_diff_from_mean"
FROM htru2, pulsar_avgs
)
SELECT SQRT(SUM(squared_diff_from_mean)/(COUNT(*))) AS "stdev_itg_mean"
FROM squared_diff
'''

avgs_df = pd.read_sql(avgs, engine)

avgs_df.head()
```
